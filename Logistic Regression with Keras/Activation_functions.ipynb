# The Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
  """
  Applies the sigmoid activation function.

  Args:
    x: Input value or array.

  Returns:
    The sigmoid of the input.
  """
  return 1 / (1 + np.exp(-x))

# Generate some data
x = np.linspace(-100, 100, 10000)  # 100 evenly spaced points between -10 and 10
y = sigmoid(x)

# Plot the sigmoid function
plt.plot(x, y,c='r')
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid Function")
plt.grid(True)
plt.show()
```

# The tanH function

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
  """
  Applies the hyperbolic tangent (tanh) activation function.

  Args:
    x: Input value or array.

  Returns:
    The tanh of the input.
  """
  return np.tanh(x)

# Generate some data
x = np.linspace(-10, 10, 100)  # 100 evenly spaced points between -10 and 10
y = tanh(x)

# Plot the tanh function
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Tanh Function")
plt.grid(True)
plt.show()
```

# The relu function

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  """
  Applies the ReLU (Rectified Linear Unit) activation function.

  Args:
    x: Input value or array.

  Returns:
    The ReLU of the input.
  """
  return np.maximum(0, x)

# Generate some data
x = np.linspace(-10, 10, 100)  # 100 evenly spaced points between -10 and 10
y = relu(x)

# Plot the ReLU function
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("relu(x)")
plt.title("ReLU Function")
plt.grid(True)
plt.show()
```

# The leaky Relu

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.02):
  """
  Applies the Leaky ReLU activation function.

  Args:
    x: Input value or array.
    alpha: Slope for negative values.

  Returns:
    The Leaky ReLU of the input.
  """
  return np.maximum(alpha * x, x)

# Generate some data
x = np.linspace(-10, 10, 100)  # 100 evenly spaced points between -10 and 10
y = leaky_relu(x)

# Plot the Leaky ReLU function
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("leaky_relu(x)")
plt.title("Leaky ReLU Function")
plt.grid(True)
plt.show()
```
