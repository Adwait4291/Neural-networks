```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set()
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense , Input, Flatten
```

```python
(X_train,y_train),(X_test,y_test)= keras.datasets.mnist.load_data()
```

```python
# Normalize the Data
X_train,X_test = X_train/255.0, X_test/255.0
```

```python
def my_ann(): 
        # pass all the layers as list inside the Sequential layer
        model= Sequential([
        Input(shape=(28,28)),
        Flatten(),
        Dense(units=512,activation='relu'),
        Dense(units=512,activation='relu'),
        Dense(units=512,activation='relu'),
        Dense(units=512,activation='relu'),
        Dense(units=10,activation='softmax')
        ])
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.02),\
                      loss=keras.losses.SparseCategoricalCrossentropy(),\
                      metrics=['accuracy'])
        return model
```

```python
mnist= my_ann()
```

```python
history=mnist.fit(X_train,y_train,validation_split=0.2,epochs=20,batch_size= 64)
```

# Model Evaluation and Prediction

```python
df= pd.DataFrame(history.history)
df.head()
```

```python
plt.plot(df.accuracy,label='Training acc')
plt.plot(df.val_accuracy,label='Val acc')
plt.legend();
```

```python
plt.plot(df.loss,label='Training loss')
plt.plot(df.val_loss,label='Val loss')
plt.legend();
```
